### Загрузим необходимые библиотеки

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Загружаем датасет с информацией о пассажирах Титаника

```python
df = pd.read_csv('https://raw.githubusercontent.com/amueller/scipy-2017-sklearn/091d371/notebooks/datasets/titanic3.csv')
```

### Смотрим на первые 10 записей из датафрейма

```python
df.head(10)
```

### Посчитаем количество выживших

```python
df['survived'].value_counts()
```

### Посмотрим величину максимальной цены за билет

```python
df['fare'].max()
```

### Посмотрим всю информацию об этих пассажирах

```python
df[df['fare'] == df['fare'].max()]
```

### Посмотрим величину минимальной цены за билет

```python
df['fare'].min()
```

### Посмотрим всю информацию об этих пассажирах

```python
df[df['fare'] == df['fare'].min()]
```

### Распределение возраста. Нормальное распределение

```python
plt.subplots(figsize = (22,10),)
df = df[df['age'].notna()]
sns.distplot(df['age'], bins = 100, kde = True, rug = False, norm_hist=False);
```

### Распеделение платы за билет. Экспоненциальное распределение

```python
plt.subplots(figsize = (22,10),)
df = df[df['fare'].notna()]
sns.distplot(df['fare'], bins = 100, kde = True, rug = False, norm_hist=False);
```

### Распределение количества родственников на борту. Экспоненциальное распределение

```python
plt.subplots(figsize = (22,10),)
sns.distplot(df['parch'], kde = True, rug = False, norm_hist=False);
```

### Распределение порядкового номера пассажира. Равномерное распределение

```python
plt.subplots(figsize = (22,10),)
sns.distplot(df.index, bins = 100, kde = True, rug = False, norm_hist=False);
```
